```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

spotifydata = pd.read_csv("spotifydata.csv")
genredata = pd.read_csv("data_by_genres.csv")
genredata
```

```python
genredata.describe()
popgenres = genredata[(genredata["genres"] == "r&b") | (genredata["genres"] == "rap") | (genredata["genres"] == "alternative rock") | (genredata["genres"] == "edm") | (genredata["genres"] == "indie pop")]
popgenres["genres"]
colorscheme = {'alternative rock':'red', 'edm':'orange', 'indie pop':'fuchsia', 'r&b':'green', 'rap':'blue'}
labels = list(colorscheme.keys())

fig, ax = plt.subplots(2, 4, figsize = (30, 15))
ax[0, 0].bar(popgenres["genres"], popgenres["acousticness"], color=colorscheme.values())
ax[0, 0].set_title("Avg Acousticness Per Genre", fontsize = 18)
ax[0, 1].bar(popgenres["genres"], popgenres["danceability"], color=colorscheme.values())
ax[0, 1].set_title("Avg Danceability Per Genre", fontsize = 18)
ax[0, 2].bar(popgenres["genres"], popgenres["valence"], color=colorscheme.values())
ax[0, 2].set_title("Avg Valence Per Genre", fontsize = 18)
ax[0, 3].bar(popgenres["genres"], popgenres["energy"], color=colorscheme.values())
ax[0, 3].set_title("Avg Energy Per Genre", fontsize = 18)
ax[1, 0].bar(popgenres["genres"], popgenres["speechiness"], color=colorscheme.values())
ax[1, 0].set_title("Avg Speechiness Per Genre", fontsize = 18)
ax[1, 1].bar(popgenres["genres"], popgenres["instrumentalness"], color=colorscheme.values())
ax[1, 1].set_title("Avg Instrumentalness Per Genre", fontsize = 18)
ax[1, 2].bar(popgenres["genres"], popgenres["duration_ms"], color=colorscheme.values())
ax[1, 2].set_title("Avg Duration Per Genre", fontsize = 18)
ax[1, 3].bar(popgenres["genres"], popgenres["popularity"], color=colorscheme.values())
ax[1, 3].set_title("Avg Popularity Per Genre", fontsize = 18)
       
# plt.Rectangle((0,0),1,1
# [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
handles = [plt.Rectangle((0,0),1,1, color=colorscheme[l]) for l in labels]
ax[0, 0].legend(handles, labels)
```
